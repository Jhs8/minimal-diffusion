```python

    
import matplotlib.pyplot as plt
import numpy as np

# 数据输入
data = []
for item in ["32,60.78,60.25", "64,81.58,80.04", "128,86.86,84.35"]:
    imgs, val_acc, test_acc = item.split(',')
    data.append((int(imgs), float(val_acc), float(test_acc)))

# 设置画布
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制折线图
imgs, val_accs, test_accs = zip(*data)
ax.plot(imgs, val_accs, marker='o', label='Validation Accuracy')
ax.plot(imgs, test_accs, marker='s', label='Test Accuracy')
# ax.scatter(50000,91.56, color='r', marker='x', label='Validation Accuracy(Original)')
# ax.scatter(50000,91.34, color='b', marker='s', label='Test Accuracy(Original)')
# ax.scatter(50000,82.80, color='g', marker='x', label='Validation Accuracy(mix)')
# ax.scatter(50000,87.81, color='black', marker='s', label='Test Accuracy(mix)')
# 设置坐标轴标签
ax.set_xlabel('Unet Size')
ax.set_ylabel('Accuracy (%)')
ax.set_xticks(imgs)

# 设置图例
ax.legend(loc='lower right')

# 调整坐标轴范围
ax.set_xlim(left=0, right=max(imgs) * 1.2)
ax.set_ylim(bottom=55, top=92)
# 设置坐标轴标签

ax.set_xticklabels(imgs, rotation=45)

# 显示网格
ax.grid(linestyle='--')
# 显示每个点的纵坐标
for x, y in zip(imgs, val_accs):
    ax.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

for x, y in zip(imgs, test_accs):
    ax.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')


# 显示图表
plt.show()
```
